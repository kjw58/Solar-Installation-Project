```python
#Webscrabing Weather Data from Current Results Website - average-annual-state-sunshine

#Load the URL of interest.
data_url = 'https://www.currentresults.com/Weather/US/average-annual-state-sunshine.php'
print(data_url)

# Request the page and use BeautifulSoup to extract the contents
page = requests.get(url = data_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Request the page and use BeautifulSoup to extract the headings
table_header = soup.find('thead')
table_headers = table_header.find_all('th')
headings_list = []
for header in table_headers:
    headings_list.append(header.text)

# Request the page and use BeautifulSoup to extract the Table of interest
#There area total of 3 tables on the page.
table = soup.find('tbody')
rows = table.find_all('tr')

#Covert the collected data into a dataframe
r = len(rows)
c = len(table_headers)
arr = [[0 for i in range(c)] for j in range(r)] 
for row in range(r):
    for col in range(c):
        value_row = rows[row].find_all('td')
        value_col = value_row[col].text
        arr[row][col] = value_col

df1 = pd.DataFrame.from_dict(arr) 
df1.columns = headings_list

# Request the page and use BeautifulSoup to extract the Table of interest
#There area total of 3 tables on the page.
tables = soup.find_all(attrs={'class':'articletable tablecol-1-left tablecol-2-left smalltext'})
#print(len(tables))
table = tables[0].find('tbody')
rows_2 = table.find_all('tr')

r2 = len(rows_2)
c2 = len(headings_list)
arr2 = [[0 for i in range(c2)] for j in range(r2)]

for row2 in range(r2):
    for col2 in range(c2):
        value_row_2 = rows_2[row2].find_all('td')
        value_col_2 = value_row_2[col2].text
        arr2[row2][col2] = value_col_2
df2 = pd.DataFrame.from_dict(arr2) 
df2.columns = headings_list

# Request the page and use BeautifulSoup to extract the Table of interest
#There area total of 3 tables on the page.
tables = soup.find_all(attrs={'class':'articletable tablecol-1-left tablecol-2-left smalltext'})
table = tables[1].find('tbody')
rows_3 = table.find_all('tr')

r3 = len(rows_3)
c3 = len(headings_list)
arr3 = [[0 for i in range(c3)] for j in range(r3)]

for row3 in range(r3):
    for col3 in range(c3):
        value_row_3 = rows_3[row3].find_all('td')
        value_col_3 = value_row_3[col3].text
        arr3[row3][col3] = value_col_3
df3 = pd.DataFrame.from_dict(arr3) 
df3.columns = headings_list

#Average Annual Sunshine combine all 3 tables into one.
frames = [df1, df2, df3]
result = pd.concat(frames)
Average_Annual_Sunshine = result
Average_Annual_Sunshine['State'] = Average_Annual_Sunshine['State'].map(us_state_abbrev)
Average_Annual_Sunshine
```

```python
#Webscrabing Weather Data from Current Results Website -average-annual-state-temperatures

#Load the URL of interest.
data_url = 'https://www.currentresults.com/Weather/US/average-annual-state-temperatures.php'
print(data_url)

# Request the page and use BeautifulSoup to extract the contents
page = requests.get(url = data_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Request the page and use BeautifulSoup to extract the Table of interest
#There area total of 3 tables on the page.
tables = soup.find_all(attrs={'class':'articletable tablecol-1-left'})

# Request the page and use BeautifulSoup to extract the header
table_header = soup.find('thead')
table_headers = table_header.find_all('th')
headings_list = []
for header in table_headers:
    headings_list.append(header.text)
# Request the page and use BeautifulSoup to extract values and place them into a dataframe
Average_Annual_Temperature = pd.DataFrame()
for i in range(len(tables)):
    table = tables[i].find('tbody')
    rows_3 = table.find_all('tr')
    r3 = len(rows_3)
    c3 = len(headings_list)
    arr3 = [[0 for i in range(c3)] for j in range(r3)]

    for row3 in range(r3):
        for col3 in range(c3):
            value_row_3 = rows_3[row3].find_all('td')
            value_col_3 = value_row_3[col3].text
            arr3[row3][col3] = value_col_3
    df = pd.DataFrame.from_dict(arr3)
    df.columns = headings_list
    Average_Annual_Temperature = pd.concat([Average_Annual_Temperature,df])
#Combine all 3 tables into one.
Average_Annual_Temperature['State'] = Average_Annual_Temperature['State'].map(us_state_abbrev)
Average_Annual_Temperature
```

```python
#Webscrabing Weather Data from Current Results Website - Average Yearly Snowfall by American State

#Load the URL of interest.
data_url = 'https://www.currentresults.com/Weather/US/average-snowfall-by-state.php'
page = requests.get(url = data_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Request the page and use BeautifulSoup to extract the Table of interest
#There area total of 3 tables on the page.
tables = soup.find_all(attrs={'class':'articletable tablecol-2-left tablecol-3-left smalltext'})


# Request the page and use BeautifulSoup to extract the header
table_header = soup.find('thead')
table_headers = table_header.find_all('th')
headings_list = []
for header in table_headers:
    headings_list.append(header.text)
    
#Extract the information from the table values and place them into Dataframes
Average_Yearly_Snowfall = pd.DataFrame()
for i in range(len(tables)):
    table = tables[i].find('tbody')
    rows_3 = table.find_all('tr')
    r3 = len(rows_3)
    c3 = len(headings_list)
    arr3 = [[0 for i in range(c3)] for j in range(r3)]

    for row3 in range(r3):
        for col3 in range(c3):
            value_row_3 = rows_3[row3].find_all('td')
            value_col_3 = value_row_3[col3].text
            arr3[row3][col3] = value_col_3
    df1 = pd.DataFrame.from_dict(arr3) 
    df1.columns = headings_list
    Average_Yearly_Snowfall = pd.concat([Average_Yearly_Snowfall,df1])

#Combine all 3 tables into one.
Average_Yearly_Snowfall['State'] = Average_Yearly_Snowfall['State'].map(us_state_abbrev)
Average_Yearly_Snowfall
```

```python
#Outer merge all three data sets
Weather_data = Average_Annual_Sunshine.merge(Average_Yearly_Snowfall,\
                            how='outer',\
                            on=['State'],\
                            )

Weather_data = Weather_data.merge(Average_Annual_Temperature,\
                            how='outer',\
                            on='State',\
                            )
```
